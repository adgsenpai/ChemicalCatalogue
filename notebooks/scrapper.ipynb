```python
from pptx import Presentation
import pandas as pd

# Open PulpanPaperDataOnly.pptx
prs = Presentation('./data/sugarproductionpartb.pptx')
data = []

for i in range(len(prs.slides)):
    slide = prs.slides[i]
    title = slide.shapes.title.text        
    if slide.shapes[0].text == 'General':
        equipment = slide.shapes[0].text.split(',')
    else:
        equipment = slide.shapes[1].text.split(',')
    # remove cont. and also AND words
    equipment = [item.replace('cont.', '') for item in equipment]
    equipment = [item.replace('and', ',') for item in equipment]
    equipment = [item.replace('/', ',') for item in equipment]    
    # account for ()
    equipment = [item.replace('(', '') for item in equipment]
    equipment = [item.replace(')', '') for item in equipment]

    # remove spaces in front of , / or and
    equipment = [item.strip() for item in equipment]
    # remove / with ,
    equipment = [item.lstrip() for item in equipment]
    equipment = [item.replace(' , ', ",") for item in equipment]
    # check if text got , in it then seperate it
    equipment = [item.lstrip() for item in equipment]
    
    
    for item in equipment:
        # if item has ,
        if ',' in item:
            # split it
            equipment.extend(item.split(','))
            # remove the item with ,
            equipment.remove(item)         

    # Initialize a variable for the table
    table = None

    # Iterate through shapes to find a table
    for shape in slide.shapes:
        if shape.has_table:
            table = shape.table
            break


    if table is not None: 
        for i, row in enumerate(table.rows):
            if i == 0:
                continue
            else:             
                row_data = {
                    'Industry': title,
                    'Equipment': equipment,        
                    'Component Part': row.cells[0].text.strip(),
                    'Lube Spec': row.cells[1].text.strip(),
                    'Lube Requirements': row.cells[2].text.strip(),
                    'Standard': row.cells[3].text.strip(),
                    'Premium': row.cells[4].text.strip(),
                    'Supreme': row.cells[5].text.strip(),
                }
                if (row_data['Component Part'] == '') and (row_data['Lube Spec'] == '') and (row_data['Lube Requirements'] == ''):
                    continue
                else:
                    data.append(row_data)       
               

# Now, 'data' contains the extracted information without duplicates
data
```

```python
# data
df = pd.DataFrame(data)
df
```

```python

# save to PulpAndPaperJSON
df.to_json('./SugarcaneProductionB.json', orient='records')
```
